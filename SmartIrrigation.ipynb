```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.multioutput import MultiOutputClassifier

from sklearn.preprocessing import MinMaxScaler
import joblib
```

```python
df = pd.read_csv("/content/68544eadb00637051626.csv")
```

```python

df =df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df.describe().T #Statics of Dataset
```

```python
x = df.iloc[:, 0:20] #This gives you columns 0 to 19 (Sensor_0 to sensor_19)

y = df.iloc[:, 20:]
```

```python
y.sample(10)
```

```python
x.info()
```

```python
y.info()
```

```python
x.shape, y.shape
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=4,
    min_samples_leaf=2,
    max_features='sqrt',
    random_state = 42
)

model = MultiOutputClassifier(rf)

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=y.columns))
```

```python
import matplotlib.pyplot as plt

# Assuming df has boolean columns: 'parcel_0', 'parcel_1', 'parcel_2'
conditions = {
    "Parcel 0 ON": df['parcel_0'],
    "Parcel 1 ON": df['parcel_1'],
    "Parcel 2 ON": df['parcel_2'],
    "Parcel 0 & 1 ON": df['parcel_0'] & df['parcel_1'],
    "Parcel 0 & 2 ON": df['parcel_0'] & df['parcel_2'],
    "Parcel 1 & 2 ON": df['parcel_1'] & df['parcel_2'],
    "All Parcels ON": df['parcel_0'] & df['parcel_1'] & df['parcel_2'],
}

fig, axs = plt.subplots(nrows=len(conditions), figsize=(10, 15), sharex=True)

for ax, (title, condition) in zip(axs, conditions.items()):
    ax.step(df.index, condition.astype(int), where='post', linewidth=1, color='teal')
    ax.set_title(f"Sprinkler - {title}")
    ax.set_ylabel("Status")
    ax.set_yticks([0, 1])
    ax.set_yticklabels(['OFF', 'ON'])

# Add x-axis label to the last subplot
axs[-1].set_xlabel("Time Index (Row Number)")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df has binary (0/1) columns: 'pump_0', 'pump_1', 'pump_2', 'parcel_0', 'parcel_1', 'parcel_2'
any_pump_on = (df['parcel_0'] == 1) | (df['parcel_1'] == 1) | (df['parcel_2'] == 1)

plt.figure(figsize=(15, 5))

# Plot each parcel's status
plt.step(df.index, df['parcel_0'], where='post', linewidth=2, label='Parcel 0 Pump', color='blue')
plt.step(df.index, df['parcel_1'], where='post', linewidth=2, label='Parcel 1 Pump', color='green')
plt.step(df.index, df['parcel_2'], where='post', linewidth=2, label='Parcel 2 Pump', color='red')

# Optional: Plot overall pump activity
plt.step(df.index, any_pump_on.astype(int), where='post', linewidth=2, label='Any Pump ON', color='orange', linestyle='--')

plt.title("Pump Activity and Combined Form Coverage")
plt.xlabel("Time Index (Row Number)")
plt.ylabel("Status")
plt.yticks([0, 1], ['OFF', 'ON'])
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
